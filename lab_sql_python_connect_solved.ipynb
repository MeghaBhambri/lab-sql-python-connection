```python
# python function called rental months

def rental_months(month:int , year: int,engine= 'mysql+pymysql://root:'+password+'@localhost/sakila'):
   
    query = f""" SELECT * from rental
    where month(rental_date) = {month} and year(rental_date) = {year}
    
    """
    data = pd.read_sql_query(query, engine)
    return (data.head(25))
```

Create a Python function called compare_rentals that takes two DataFrames as input containing the number of rentals made by each customer in different months and years. The function should return a combined DataFrame with a new 'difference' column, which is the difference between the number of rentals in the two months.

```python
def compare_rentals(d1:pd.DataFrame, d2: pd.DataFrame):
    display(d1)
    if d1 is not None and d2 is not None:
        d3 = pd.merge(d1,d2,on='customer_id', how='outer')
        if not d3.empty:
            month1 = d1.columns[0]
            month2 = d2.columns[0]
            d3['difference'] = abs((d3[month1] - d3[month2])).fillna(0)
            return d3
        else:
            print("One or both DataFrames are empty.")
            return None
    else:
        print("One or both of the input DataFrames are None.")
        return None

    return (d3)
```
